dask array로 받기

```python
ruta ='...'
x = df6[['내부 온도 1 평균',	'내부 온도 1 최고']]

def transf(x):
    xd=x.to_delayed()
    # full = [dask.array.from_delayed(i, i.compute().shape, i.compute().dtype) for i in xd]
    #데이터프레임.dtype가 안먹서 일단 object로 해봄
    full = [dask.array.from_delayed(i, i.compute().shape,float) for i in xd]
    return dask.array.concatenate(full)

x_array=transf(x)
```
